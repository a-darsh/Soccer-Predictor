```python
# Import statements

import pandas as pd
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler
from joblib import dump
from joblib import load
```

```python
# Loading training Data
data = pd.read_csv('train_data.csv')
data
```

```python

# Handle missing values
data.fillna(method='ffill', inplace=True)


# Taking necessary columns
data = data[['HomeTeam_PositiveSentiment','HomeTeam_NeutralSentiment','HomeTeam_NegativeSentiment','AwayTeam_PositiveSentiment','AwayTeam_NeutralSentiment','AwayTeam_NegativeSentiment','AvgOdds_HomeWin','AvgOdds_Draw','AvgOdds_AwayWin','HomeGoals','AwayGoals']]

# Normalize or standardize the numerical features
scaler = StandardScaler()
numerical_cols = ['HomeTeam_PositiveSentiment', 'HomeTeam_NeutralSentiment', 'HomeTeam_NegativeSentiment',
                  'AwayTeam_PositiveSentiment', 'AwayTeam_NeutralSentiment', 'AwayTeam_NegativeSentiment',
                  'AvgOdds_HomeWin', 'AvgOdds_Draw', 'AvgOdds_AwayWin']
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Feature Engineering
# You can add new features if needed

# Model Building
X = data.drop(['HomeGoals', 'AwayGoals'], axis=1)
y_home = data['HomeGoals']
y_away = data['AwayGoals']

# Split the data
X_train, X_test, y_home_train, y_home_test, y_away_train, y_away_test = train_test_split(
    X, y_home, y_away, test_size=0.2, random_state=42)

```

```python
X_train
```

```python
# Initialize and train the models
model_home = XGBRegressor()
model_away = XGBRegressor()

# Model Training
model_home.fit(X_train, y_home_train)
model_away.fit(X_train, y_away_train)


```

```python
# Model Evaluation
y_home_pred = model_home.predict(X_test)
y_away_pred = model_away.predict(X_test)

rmse_home = mean_squared_error(y_home_test, y_home_pred, squared=False)
rmse_away = mean_squared_error(y_away_test, y_away_pred, squared=False)

print(f'RMSE for Home Goals: {rmse_home}')
print(f'RMSE for Away Goals: {rmse_away}')
```

```python
# Model saving
dump(model_home, 'model_home.joblib')
dump(model_away, 'model_away.joblib')
dump(scaler, 'scaler.joblib')
```

```python
# Prediction Function

def predict_match_result(input_data, model_home, model_away, scaler):
    """
    Predicts Home Goals, Away Goals, and the match result.

    Parameters:
    input_data (dict): Match data containing features but not the goals.
    model_home (model): Trained model for predicting Home Goals.
    model_away (model): Trained model for predicting Away Goals.
    scaler (StandardScaler): Scaler used for standardizing features.
    label_encoder (LabelEncoder): Encoder used for categorical variables.

    Returns:
    dict: Predicted Home Goals, Away Goals, and match result.
    """

    # Convert input data to DataFrame
    input_df = pd.DataFrame([input_data])
    home_team = input_df['Home'].values[0]
    print(home_team)
    away_team = input_df['Away'].values[0]

    # Handle missing values
    input_df = input_df[['HomeTeam_PositiveSentiment','HomeTeam_NeutralSentiment','HomeTeam_NegativeSentiment','AwayTeam_PositiveSentiment','AwayTeam_NeutralSentiment','AwayTeam_NegativeSentiment','AvgOdds_HomeWin','AvgOdds_Draw','AvgOdds_AwayWin']]

    # Standardize numerical features
    numerical_cols = ['HomeTeam_PositiveSentiment', 'HomeTeam_NeutralSentiment', 'HomeTeam_NegativeSentiment',
                      'AwayTeam_PositiveSentiment', 'AwayTeam_NeutralSentiment', 'AwayTeam_NegativeSentiment',
                      'AvgOdds_HomeWin', 'AvgOdds_Draw', 'AvgOdds_AwayWin']
    input_df[numerical_cols] = scaler.transform(input_df[numerical_cols])

    # Predict Home and Away Goals
    predicted_home_goals = round(model_home.predict(input_df)[0])
    predicted_away_goals = round(model_away.predict(input_df)[0])


    # Determine match result
    if predicted_home_goals > predicted_away_goals:
        result = 'Home Win'
    elif predicted_home_goals < predicted_away_goals:
        result = 'Away Win'
    else:
        result = 'Draw'

    return {
        'Predicted Home Goals': predicted_home_goals,
        'Predicted Away Goals': predicted_away_goals,
        'Match Result': result,
        'Home Team': home_team,
        'Away Team': away_team
    }


```

```python
# Loading the model
model_home = load('model_home.joblib')
model_away = load('model_away.joblib')
scaler = load('scaler.joblib')
```

```python

# Example usage
match_data = {
    'Date': '2023-11-16',
    'Home': 'Liverpool',
    'Away': 'Man City',
    'HomeTeam_PositiveSentiment': 0.9,
    'HomeTeam_NeutralSentiment': 0.2,
    'HomeTeam_NegativeSentiment': 0.1,
    'AwayTeam_PositiveSentiment': 0.6,
    'AwayTeam_NeutralSentiment': 0.3,
    'AwayTeam_NegativeSentiment': 0.1,
    'AvgOdds_HomeWin': 1.5,
    'AvgOdds_Draw': 1.2,
    'AvgOdds_AwayWin': 6.8
}
predictions = predict_match_result(match_data, model_home, model_away, scaler)
print(predictions)
```
